```python
import pandas as pd
import numpy as np
import scipy as sp
import scipy.cluster.hierarchy as hr
import matplotlib.pyplot as plt
import sklearn.datasets as sk_data
import sklearn.metrics as metrics
from sklearn.cluster import KMeans
from scipy.spatial.distance import pdist, squareform
import seaborn as sns
%matplotlib inline
```

```python
# Preprocessed the data into json files.
# This was done in a separate file called preprocess.py

# list of all countries and their designated cuisine type
countries = pd.read_json("scirep-cuisines-detail/country.json")

# list of all recipes, containing their ingredients and country
allr = pd.read_json("scirep-cuisines-detail/allr.json")
allr = allr.fillna(value=0)
allr.head(10)
```

```python
country_recipes = allr.groupby("country")
```

```python
# Generate a sample of 500 to work on.
# I don't know the data at all, nor do I have enough experience to
# have a of frame of reference for how long certain data mining
# techniques will take. So it isn't feasible to "get to know my data"
# by randomly trying various techniques on such a large data set without
# some sort of estimation of how long each will take. A smaller sample
# set will at least give me results in sub 60seconds.
sample_size = 500
def takeSample(data, size):
    return data.loc[np.random.choice(data.index, size, replace=False)]

sample = takeSample(allr, sample_size)

# Used for removing non-numeric attributes. In this case, just the
# 'country' attribute.
metricColumns = [column for column in allr.columns if allr[column].dtype == 'float64']

sns.heatmap(sample[metricColumns], xticklabels=False, yticklabels=False, linewidths=0,cbar=True)
```

```python
euclidean_dists = metrics.euclidean_distances(sample[metricColumns])
```

```python
sns.heatmap(euclidean_dists, xticklabels=False, yticklabels=False, linewidths=0, square=True, cbar=True)
```

```python
def evaluate_clusters(X,max_clusters):
    error = np.zeros(max_clusters+1)
    error[0] = 0;
    for k in range(1,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        error[k] = kmeans.inertia_

    plt.plot(range(1,len(error)),error[1:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Error')
```

```python
evaluate_clusters(sample[metricColumns], 25)
```

```python
kmeans = KMeans(init='k-means++', n_clusters=15, n_init=10)
kmeans.fit_predict(sample[metricColumns])
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_

print "The total error of the clustering is: ", error
print '\nCluster labels'
print labels
print '\n Cluster Centroids'
print centroids
```

```python
idx = np.argsort(labels)
rearranged_dists = euclidean_dists[idx,:][:,idx]
sns.heatmap(rearranged_dists, xticklabels=False, yticklabels=False, linewidths=0, square=True, cbar=True)
```

```python
# Nothing particularly interesting is showing up.
# The goal was to see if certain countries had similar
# ingredients used in their recipes. Let's print out
# each ingredient and how many recipes it is used in.
# Maybe it's a classic example of the 80/20 case.

ingredient_count = []
for c in allr[metricColumns].columns:
    ingredient_count.append({"ingredient": c, "count": allr[c].sum()})
    
ingredient_count = pd.DataFrame(ingredient_count).sort(columns="count", ascending=False).set_index("ingredient")


ingredient_count.plot(figsize=(10, 10))
ingredient_count.head(75).plot(kind="bar", figsize=(20, 10))
```

```python
# Graphs made it clear the vast majority of ingredients are only used
# in a fraction of the recipes. Let's look at which countries most
# frequently use each ingredient.

# top 20 ingredients
top_ingredients = ingredient_count.head(20).index

for ingredient in top_ingredients:
    pd.value_counts(allr[allr[ingredient] == 1]["country"]).plot(kind="bar")
    plt.title(ingredient)
    plt.show()
```

```python
# "american" cuisine uses unusually large amounts.
# Lets see what things look like when we ignore it.
# Hopefully things look more interesting and there is
# more variety.

no_american = allr[allr.country != "american"].copy()
for ingredient in top_ingredients:
    pd.value_counts(no_american[no_american[ingredient] == 1]["country"]).plot(kind="bar")
    plt.title(ingredient)
    plt.show()
```

```python
sample = takeSample(no_american, sample_size)

euclidean_dists = metrics.euclidean_distances(sample[metricColumns])
sns.heatmap(euclidean_dists, xticklabels=False, yticklabels=False, linewidths=0, square=True, cbar=True)
```

```python
evaluate_clusters(sample[metricColumns], 25)
```

```python
evaluate_clusters(sample[metricColumns], 10)
```

```python
idx = np.argsort(labels)
rearranged_dists = euclidean_dists[idx,:][:,idx]
sns.heatmap(rearranged_dists, xticklabels=False, yticklabels=False, linewidths=0, square=True, cbar=True)
```

```python
# Slightly more interesting. You can see some slight signs
# of general clustering. Looks like it might even be the case
# that each country has its own unique pairing of ingredients.
# Might be worth exploring. Perhaps classification.
#
# There is around 35 or so unique countries listed. Perhaps
# a higher cluster count will show a significant drop in error.

evaluate_clusters(sample[metricColumns], 35)
```

```python
kmeans = KMeans(init='k-means++', n_clusters=5, n_init=35)
kmeans.fit_predict(sample[metricColumns])
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_

idx = np.argsort(labels)
rearranged_dists = euclidean_dists[idx,:][:,idx]
sns.heatmap(rearranged_dists, xticklabels=False, yticklabels=False, linewidths=0, square=True, cbar=True)
```

```python
# Lets give hierarchically clustering a whirl.
sample = takeSample(no_american, 100)

Z = hr.linkage(sample[metricColumns], method='complete', metric='euclidean')

print Z.shape
```

```python
fig = plt.figure(figsize=(20,10))
T = hr.dendrogram(Z,color_threshold=0.4, leaf_font_size=4)
fig.show()
```

```python
distanceMatrix = pdist(sample[metricColumns],'euclidean')
```

```python
ck = hr.fcluster(hr.linkage(distanceMatrix, method='complete'), 1.5,'distance')
print ck
```
